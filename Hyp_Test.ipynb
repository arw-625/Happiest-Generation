```python
import numpy as np
import scipy as sp
import scipy.stats as stats
import matplotlib.pyplot as plt
```

```python
binomial = stats.binom(n=2300, p=0.28)
binomial_mean = 0.28 * 2300
binomial_var = 2300 * 0.28 * 0.72
normal_approx = stats.norm(binomial_mean, np.sqrt(binomial_var))
x = np.linspace(0, 2300, num=2400)

fig, ax = plt.subplots(figsize=(16, 6))

bar_sizes = [binomial.pmf(i) for i in range(2300)]
bars = ax.bar(range(2300), bar_sizes, align="center")
ax.plot(x, normal_approx.pdf(x))
ax.set_xlim(525,750)
plt.savefig("images/hyp_dist")
#ax.set_title("# of Kickflips Landed Under The Null Hypothesis");
```

```python
p_value = 1 - normal_approx.cdf(636)
print("p-value for 10 year suicide experiment: {:2.2f}".format(p_value))
```

```python
fig, ax = plt.subplots(1, figsize=(16, 3))

ax.plot(x, normal_approx.pdf(x), linewidth=3)
ax.set_xlim(525,750)
ax.fill_between(x, normal_approx.pdf(x), 
                where=(x >= 636), color="red", alpha=0.5)
ax.set_title("p-value Region");
plt.savefig("images/p_val")
#Failed to reject the null hypothesis
```
